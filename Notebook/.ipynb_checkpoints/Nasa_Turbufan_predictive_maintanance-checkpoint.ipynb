```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
```

```python

base_path = r'D:\project\RUL\data' 


index_names = ['engine', 'cycle']
setting_names = ['setting1', 'setting2', 'setting3']
sensor_names = [f'sensor_{i}' for i in range(1, 22)]
col_names = index_names + setting_names + sensor_names


df_train = pd.read_csv(os.path.join(base_path, 'train_FD001.txt'), sep=r'\s+', header=None, names=col_names)
df_test = pd.read_csv(os.path.join(base_path, 'test_FD001.txt'), sep=r'\s+', header=None, names=col_names)
rul = pd.read_csv(os.path.join(base_path, 'RUL_FD001.txt'), sep=r'\s+', header=None, names=['RUL']).values.flatten()

    
```

```python
max_cycles = df_train.groupby("engine")["cycle"].max().reset_index()
max_cycles.columns = ["engine", "max_cycle"]

df_train = df_train.merge(max_cycles, on="engine", how="left")
df_train["RUL"] = df_train["max_cycle"] - df_train["cycle"]
df_train.drop(columns=["max_cycle"], inplace=True)
```

```python
print(df_train.columns.tolist())
```

```python
constant_cols = [col for col in df_train.columns if df_train[col].nunique() == 1]
print("Columns with constant values:", constant_cols)
df_train.drop(constant_cols, axis=1, inplace=True, errors='ignore')
df_test.drop(constant_cols, axis=1, inplace=True, errors='ignore')
```

```python
df_train_RUL = df_train.groupby(['engine']).agg({'cycle':'max'})
df_train_RUL.rename(columns={'cycle':'life'},inplace=True)
df_train_RUL.head()
```

```python
RUL_THRESHOLD = 120
df_train['RUL'] = df_train['RUL'].clip(upper=RUL_THRESHOLD)
df_train['RUL'].describe()
```

```python
engine_life = df_train.groupby('engine')['cycle'].max()
print("Mean lifespan:", engine_life.mean())
print("Min lifespan :", engine_life.min())
print("Max lifespan :", engine_life.max())

sns.histplot(engine_life, kde=True)
plt.title("Engine Lifespan Distribution")
plt.show()
```

```python
corr = df_train.corr()
rul_corr = corr[['RUL']].sort_values(by='RUL', ascending=False)

plt.figure(figsize=(10, 8))
sns.heatmap(rul_corr, annot=True, cmap='RdYlGn', fmt=".2f")
plt.title("Sensor Correlation with RUL")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
features_for_pca = [
    c for c in df_train.columns
    if c not in ['engine','cycle','RUL']
]

X_scaled = StandardScaler().fit_transform(df_train[features_for_pca])

pca = PCA(n_components=3)
pcs = pca.fit_transform(X_scaled)

plt.scatter(pcs[:,0], pcs[:,1], c=df_train['RUL'], cmap='jet_r', s=10)
plt.colorbar(label="RUL")
plt.title("PCA Degradation Trajectory")
plt.show()

print("Explained variance:", pca.explained_variance_ratio_.sum())
```

```python
early_life = df_train[df_train['cycle'] < 20]
late_life = df_train[df_train['RUL'] < 20]

sensors_to_compare = ['sensor_4', 'sensor_11', 'sensor_12', 'sensor_2']
available_sensors = [s for s in sensors_to_compare if s in df_train.columns]

if available_sensors:
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 10))
    axes = axes.flatten()

    for i, sensor in enumerate(available_sensors):
        sns.kdeplot(early_life[sensor], fill=True, color="green", label='Healthy (Start)', ax=axes[i], alpha=0.3)
        sns.kdeplot(late_life[sensor], fill=True, color="red", label='Failing (End)', ax=axes[i], alpha=0.3)
        axes[i].set_title(f'{sensor}: Healthy vs Failing Separation')
        axes[i].legend()
    plt.tight_layout()
    plt.show()
else:
    print("Warning: Selected sensors were dropped during cleaning.")
```

```python
y_true=rul.flatten()
y_true = rul
y_true
```

```python
features_to_scale = features_for_pca
scaler = MinMaxScaler()
df_train[features_to_scale] = scaler.fit_transform(df_train[features_to_scale])
df_test[features_to_scale]  = scaler.transform(df_test[features_to_scale])  
X_train = df_train[features_to_scale]
y_train = df_train['RUL']
X_test = (
    df_test
    .groupby('engine')    
    .last()
    .reset_index()[features_to_scale]
)
y_true=rul.flatten()
y_true = rul
```

```python
X = df_train[features_to_scale]
y = df_train['RUL']

selector = RandomForestRegressor(n_estimators=100, random_state=42)
selector.fit(X, y)
importance    = pd.Series(selector.feature_importances_, index=features_to_scale)
top_8_sensors = importance.nlargest(8).index.tolist()
print(f"Top 8 sensors: {top_8_sensors}")
```

```python
df_train[top_8_sensors] = scaler.fit_transform(df_train[top_8_sensors])
df_test[top_8_sensors]  = scaler.transform(df_test[top_8_sensors])
```

```python

X_train = df_train[top_8_sensors].values
X_test  = df_test.groupby('engine').last().reset_index()[top_8_sensors].values
#y_true  = true_rul.flatten()

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y)
rf_pred  = rf_model.predict(X_test)
rf_rmse  = np.sqrt(mean_squared_error(y_true, rf_pred)) 
rf_r2    = r2_score(y_true, rf_pred)
print(f"RF   RMSE: {rf_rmse:.2f}  R2: {rf_r2:.2f}")

xgb_model = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.03, max_depth=8,
                               subsample=0.8, colsample_bytree=0.8, random_state=42)
xgb_model.fit(X_train, y)
xgb_pred  = xgb_model.predict(X_test)
xgb_rmse  = np.sqrt(mean_squared_error(y_true, xgb_pred))
xgb_r2    = r2_score(y_true, xgb_pred)
print(f"XGB  RMSE: {xgb_rmse:.2f}  R2: {xgb_r2:.2f}")
```

```python
import joblib

joblib.dump(xgb_model,     'model.pkl')
joblib.dump(scaler,        'scaler.pkl')
joblib.dump(top_8_sensors, 'features.pkl')

print("Saved successfully!")
```
